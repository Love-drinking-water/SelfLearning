- 括号   
1) 原始的括号   
`()`和`[]`以及`\{\}`  
2) 尖括号   
`\langle`和`\rangle`表示$\langle \rangle$   
3) 上取整   
`\lceil`和`\rceil`表示$\lceil \rceil$   
4) 下取整   
`\lfloor`和`\rfloor`表示$\lfloor \rfloor$   
5) 自适应括号    
`$\left(\frac{x}{y}\right)$`表示$\left(\frac{x}{y}\right)$  
`$\left[\frac{x}{y}\right]$`表示$\left[\frac{x}{y}\right]$   
`$\left\{\frac{x}{y}\right\}$`表示$\left\{\frac{x}{y}\right\}$  
`$\left\langle\frac{x}{y}\right\rangle$`表示$\left\langle\frac{x}{y}\right\rangle$

- 数学多行表达式   
1) **多行左括弧表达**   
使用`\\`来分行，使用`&`来表示需要对齐的位置   
```
$$
f(n)=
\begin{cases}
\cfrac n2, &if\ n\ is\ even\\
3n + 1, &if\  n\ is\ odd
\end{cases}
$$
```
$$
f(n)=
\begin{cases}
\cfrac n2, &if\ n\ is\ even\\
3n + 1, &if\  n\ is\ odd
\end{cases}
$$
2) **多行等式表达**   
```
$$
\begin{equation}\begin{split} 
a&=b+c-d \\ 
&\quad +e-f\\ 
&=g+h\\ 
& =i 
\end{split}\end{equation}
$$
```
$$
\begin{equation}\begin{split} 
a&=b+c-d \\ 
&\quad +e-f\\ 
&=g+h\\ 
& =i 
\end{split}\end{equation}
$$
其中`begin{equation}` 表示开始方程，`end{equation}` 表示方程结束；`begin{split}` 表示开始多行公式，`end{split}` 表示结束；公式中用`\\` 表示回车到下一行，`&` 表示对齐的位置。   
3) **多行方程式**    
使用`\begin{array}...\end{array}` 与`\left \{ 与\right.` 配合表示方程组:   
```
$$
\left \{ 
\begin{array}{c}
a_1x+b_1y+c_1z=d_1 \\ 
a_2x+b_2y+c_2z=d_2 \\ 
a_3x+b_3y+c_3z=d_3
\end{array}
\right.
$$
```
$$
\left \{ 
\begin{array}{c}
a_1x+b_1y+c_1z=d_1 \\ 
a_2x+b_2y+c_2z=d_2 \\ 
a_3x+b_3y+c_3z=d_3
\end{array}
\right.
$$
